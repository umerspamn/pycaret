```python
df=pd.read_csv("currencies_data.csv")
df
```

```python
thershold=0
df["target1"]=(df["percentChange24h"]>=0).astype(int)
```

```python
print(df["target1"].value_counts())
```

```python
cat=["marketPairCount", "circulatingSupply", "totalSupply", "maxSupply", "isActive"]
experiment=setup(df,target="target1",categorical_features=cat)
```

```python
best_model=compare_models()
```

```python
dfc=predict_model(best_model,df.head())
```

```python
import plotly.express as px

# Assuming 'data' is your DataFrame containing the predictions
# Replace 'data' with the actual variable name you're using

# Create a scatter plot of prediction scores
fig = px.scatter(dfc, x='prediction_score', y='target1', color='prediction_label',
                 title='Prediction Scores vs. Actual Labels')

# Add a trendline
fig.update_traces(mode='markers+lines')

# Show the plot
fig.show()
```
