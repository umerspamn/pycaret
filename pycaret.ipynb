```python
import pandas as pd
from pycaret.classification import *
```

```python
cat=["sex","cp","fbs","restecg","exang",'thal']
```

```python
experiment=setup(df,target="target",categorical_features=cat)
```

```python
predict_model(best_model,df.tail())
```
